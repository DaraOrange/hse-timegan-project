```python
from metrics.visualization_metrics import visualization
import numpy as np
import data_loading
import warnings

warnings.filterwarnings('ignore')
```

```python
def make_vis(data_name, init, seq_len):
    exp_name = data_name + ('_init' if init else '_emb')
    generated_data = np.load(f'experiments/{exp_name}_gen.npy')
    ori_data = data_loading.data_loading(data_name, seq_len)
    visualization(ori_data, generated_data, 'pca')
    visualization(ori_data, generated_data, 'tsne')
```

### Google Stock Experiment

```python
# RNN
make_vis('stock', True, 24)
```

```python
# Transformer
make_vis('stock', False, 24)
```

### Yahoo stock experiment

```python
# RNN
make_vis('yahoo_stock', True, 24)
```

```python
# Transfomer
make_vis('yahoo_stock', False, 24)
```

### Energy Experiment

```python
make_vis('energy', True, 24)
```
